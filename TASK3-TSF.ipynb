```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Superstore dataset
superstore_data = pd.read_csv("D:\data\SampleSuperstore.csv")
```

```python
# Display the first few rows of the dataset
print(superstore_data.head())

# Summary statistics of numerical columns
print(superstore_data.describe())

# Check for missing values
print(superstore_data.isnull().sum())
```

```python
superstore_data.columns
```

```python
# Sales performance by Region
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=superstore_data, estimator=sum)
plt.title('Sales Performance by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Profit', data=superstore_data)
plt.title('Profitability Analysis by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Profit')
plt.show()
```

```python
# Geographic expansion opportunities
plt.figure(figsize=(10, 6))
sns.countplot(x='State', data=superstore_data, order=superstore_data['State'].value_counts().index)
plt.title('Sales Distribution by State')
plt.xlabel('State')
plt.ylabel('Number of Sales')
plt.xticks(rotation=90)
plt.show()
```
