```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:\data\Indian Premier League\matches.csv")
```

```python
print(data.head(10))
print(data.tail(10))
data.isna().sum()
```

```python
df = pd.read_csv("D:\data\Indian Premier League\deliveries.csv")
```

```python
df
```

```python
df.columns
```

```python
# Total wickets by each bowler
bowler_wickets = df[df['player_dismissed'] != 'None'].groupby('bowler').size().sort_values(ascending=False)
print(bowler_wickets.head(10))

# Plot the top 10 bowlers by wickets
bowler_wickets.head(10).plot(kind='bar', figsize=(10, 5))
plt.title('Top 10 Bowlers by Wickets')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.show()
```

```python
# Analyze the impact of toss decision
toss_wins = data[data['toss_winner'] == data['winner']].shape[0]
toss_losses = data[data['toss_winner'] != data['winner']].shape[0]

# Plot toss decision impact
labels = ['Toss Win and Match Win', 'Toss Win and Match Loss']
sizes = [toss_wins, toss_losses]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Impact of Toss Winning on Match Outcome')
plt.axis('equal')
plt.show()
```

```python
batsman_runs = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
print(batsman_runs.head(10))

# Plot the top 10 batsmen by runs
batsman_runs.head(10).plot(kind='bar', figsize=(10, 5))
plt.title('Top 10 Batsmen by Runs')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.show()
```

```python
team_wins = data['winner'].value_counts()
print("Most successful teams")
print(team_wins)

# Plot the number of wins for each team
import matplotlib.pyplot as plt

team_wins.plot(kind='bar', figsize=(10, 5))
plt.title('Number of Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```
